<h2>Problem 1</h2>

Надо просто хранить N/B чиселок $b_i$, каждое из которое является аккумулированной суммой первых $i$ блоков.   
Тогда для ответа на запрос суммы от $a_1$ по $a_i$, нужно сложить $b_{i // B - 1} + \sum_{k=i - i\%B + 1}^{i} a_i$  
Последняя сумма считается за одно обращение к диску, как и $b_{i // B - 1}$.  
$\sum_{k=i}^j a_k = \sum_{k=1}^j a_k - \sum_{k=1}^i a_k$, так что это 4 чтения с диска.

<h2>Problem 2</h2>

Разбиваем условно все на $N/B$ блоков, и строим над ними sparse table, который будет отвечать на запросы вида "минимум на блоках $[i/B, j/B]$". Размер такой таблицы будет $N/B + 1/2 \cdot N/B +... < 2 \cdot N/B = O(N/B)$ 

Для его построения надо узнать минимумы на каждом из блоков ($O(N/B)$) для начала. Это первый ряд таблицы.  
Для построения i-го ряда таблицы надо тоже $O(N/B)$ обращений к диску. Всего таких рядов $log(N/B) < log N < B$, а значит всего обращений к диску $O(N)$. 

<h2>Problem 3</h2>

Строим дерево отрезков над начальным массивом данных, а вернее над его разбиением блоками.  
Всего $N/B$ листьев, и построить дерево можно на $O(N/B log(N/B))$.  
Итак, в вершинах хранится минимум соответствующего отрезка, а также минимум левого сына и правого (если это не лист).  

Модификация.  
Для прибавления $c$ надо прибавить его сначала для элементов в крайних блоках, обновив минимум у этих блоков. Затем надо его добавить в крайние листья каждого уровня, поднимаясь все выше, пока пути к корню не сойдутся для двух краев запросного отрезка.  
На каждом уровне затрагивается не более 4-х вершин (все остальные вершины модифицируются на более высоких уровнях). Всего уровней $log(N/B)$, операция модификации выполняется за 1 обращение к диску, а значит, операция модификации отрезка требует $O(log(N/B))$ обращений к диску.  

Взятие минимума.  
Для минимума надо подняться к максимальному общему предку, поддерживая с каждой стороны текущий минимум. На это нужно константное число операций на каждый уровень, итого $O(log(N/B))$ операций обращения к диску на взятие минимума. 

Итого сложность выполнения операций $O(M log(N/B))$. Требуемая сложность операций должна наталкивать на мысль, что нужно как-то обрабатывать их вместе, батчами. Но общего механизма для обработки батчей так, чтобы улучшить асимптотику на ум не приходит, а потому просто приму B = logN. Так вроде можно, поскольку в память logN влазит по условию. А значит, $O(M log(N/B)) = O(M/B log(N/B) logN)$. Общая сложность $O((N + M logN)/B \cdot log(N/B)) = O(((N + M) logN)/B \cdot log(N/B)) = O((N + M logN)/B \cdot log(((N + M)logN)/B)) = O(Sort((N+M)logN))$

Притянуто за уши, конечно, но что поделать.  

<h2>Problem 4</h2>

Сказано, что $V<|M|$, что я понимаю как "минимальное остовное дерево/лес поместится в память".  
В памяти предлагается держать все вершины, и закидывать пачки из V новых ребер. Буду все еще считать, что все это в память влазит (V вершин и 2V ребер).  
Далее в памяти отрабатывает алгоритм Крускала за $O(V log V)$, и так $E/V$ раз, т.е. процессорное время выполнения $O(E log V)$. При этом с диска нужно O(Scan(E + V)) чтений.  
Корректность алгоритма объясняется тем, что если бы в какой-то момент мы выкинули бы ребро из минимального остовного дерева, то это значило бы, что оно создает цикл на каком-то множестве вершин. Значит, какое-то из ребер этого цикла отсутствует в минимальном остовном дереве. Значит, заменим выкинутое ребро на отсутствующее в минимальном остовном дереве, и получим более минимальное остовное дерево.
